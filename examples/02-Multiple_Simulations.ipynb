# Run simulators in parallel

The ``ParallelSimulator`` class allows multiple simulations to be grouped and in parallel if needed.

```python
from NetworkSim.simulation.tools.plot import *
from NetworkSim.simulation.simulator.parallel import ParallelSimulator
```

## Create simulator with 10% - 40% loading

```python
simulator = ParallelSimulator(
    convergence=[True],
    average_data_rate=[10, 20, 30, 40],
    n_jobs=1
)
simulator.initialise()
```

## Run simulations

```python
simulator.run()
```

## Get summary of all simulations

```python
simulator.summary()
```

## Generate heatmap plots for simulations

```python
titles = ['$10\%$','$20\%$','$30\%$','$40\%$']
grid = [2, 2]
```

```python
delay_plot = plot_on_grid(simulator=simulator, grid=grid, titles=titles, summary_type='latency')
```

```python
packet_count_plot = plot_on_grid(simulator=simulator, grid=grid, titles=titles, summary_type='count')
```
