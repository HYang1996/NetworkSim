```python
from NetworkSim.simulation.process.ram import RAM
from NetworkSim.simulation.process.transmitter import Transmitter
from NetworkSim.simulation.process.receiver import Receiver
from NetworkSim.architecture.setup.model import Model
from NetworkSim.architecture.base.network import Network
import simpy
```

```python
env =simpy.Environment()
model = Model(network=Network(num_nodes=5))
ram0 = RAM(env=env, until=2000, ram_id=0, model=model)
ram0.initialise()
trans0 = Transmitter(env=env, ram=ram0, transmitter_id=0, model=model)
trans0.initialise()
rec2 = Receiver(env=env, receiver_id=2, model=model)
rec2.initialise()
rec4 = Receiver(env=env, receiver_id=3, model=model)
rec4.initialise()
```

```python
env.run(until=2000)
```
